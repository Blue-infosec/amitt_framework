```python
# library
import seaborn as sns
import pandas as pd
import numpy as np
 
# Create a dataset (fake)
df = pd.DataFrame(np.random.random((10,12)), columns=["a","b","c","d","e","f","g","h","i","j","k","l"])
```

```python
sns.heatmap(df, annot=True, annot_kws={"size": 7})
```

```python
sns.heatmap(df, annot=False, annot_kws={"size": 7})
```
