# Generate powerpoint slide of matrix

Generate powerpoint slide containing current version of framework

Cool thing to do, but there are other things to do first (like, for now, an html table will do). 

```python
# Load in AMITT details
import pandas as pd
arr = pd.read_csv('../matrices/matrix_arr.csv', header=None)
arr
```

```python
from pptx import Presentation
from pptx.util import Inches

prs = Presentation()
slide = prs.slides.add_slide(prs.slide_layouts[6]) # blank slide
shapes = slide.shapes

# shapes.title.text = 'AMITT'

cols = len(arr.columns)
rows = len(arr)
left = top = Inches(0.5)
width = Inches(9)
height = Inches(0.5)

table = shapes.add_table(rows, cols, left, top, width, height).table

# set column widths
# table.columns[0].width = Inches(2.0)
# table.columns[1].width = Inches(4.0)

# write column headings
table.cell(0, 0).text = 'Foo'
table.cell(0, 1).text = 'Bar'

# write body cells
table.cell(1, 0).text = 'Baz'
table.cell(1, 1).text = 'Qux'

prs.save('test.pptx')
```
